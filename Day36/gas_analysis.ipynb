# Find Outlier of Natural Gas Price in 2024

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df2024 = pd.read_csv('year2024.csv')
df2024.head()
```

```python
df2025 = pd.read_csv('year2025.csv')
df2025.head()
```

```python
stat2024 = df2024.describe().to_dict()['value']
stat2025 = df2025.describe().to_dict()['value']
```

```python
print(stat2024, '\n', stat2025)
```

```python
upper_2024 = stat2024['75%'] + 1.5 * (stat2024['75%'] - stat2024['25%'])
lower_2024 = stat2024['25%'] - 1.5 * (stat2024['75%'] - stat2024['25%'])
print('2024 upper:', upper_2024, 'lower:', lower_2024)
```

```python
sent = False
if stat2025['mean'] > upper_2024:
    sent = True
elif stat2025['mean'] < lower_2024:
    sent = True
```
